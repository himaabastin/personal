
##  Dynamic vs Static Programming Languages

- **Static Languages (e.g., C, Java):** Variable types are defined at compile time.  
- **Dynamic Languages (e.g., Python, JavaScript):** Variable types are determined at runtime.


```python

# Dynamic Typing in Python
x = 10       # integer
print(type(x))

x = "Hello"  # now a string
print(type(x))
```
